```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1986 = pd.read_clipboard(header=None)
```

```python
df_1986
```

```python
df_1986_2 = df_1986.copy()

d = df_1986_2[~df_1986_2[0].str.contains('（')]
d = d.reset_index()
d["name"] = d[0]
d = pd.DataFrame(d["name"])

for i in range(len(d)):
    if not d["name"][i][0].isdecimal():
        
        d["name"][i-1] +="を"
        d["name"][i-1] += d["name"][i]

d_lst = []

for i in range(len(d)):
    if  d["name"][i][0].isdecimal():
        
        if  d["name"][i][1].isdecimal():
            str_temp = d["name"][i][3:]
            
        if  d["name"][i][2].isdecimal():
            str_temp = d["name"][i][4:]
            
        if not d["name"][i][1].isdecimal() and not d["name"][i][2].isdecimal():
            str_temp = d["name"][i][2:]
            
        d_lst.append(str_temp)
        
df_1986_new = pd.DataFrame()
df_1986_new["name"] = d_lst

df_1986_new
```
