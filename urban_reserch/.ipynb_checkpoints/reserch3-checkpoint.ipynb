```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
from io import StringIO
from glob import glob
import os

from janome.tokenizer import Tokenizer
from janome.analyzer import Analyzer
from janome.charfilter import *
from janome.tokenfilter import *

import collections
```

```python
df_name = ["1986","1992","1998","2004","2011","2017"]

# df_lst = [pd.read_csv(f"data3_csv/{i}_new.csv", index_col=0, encoding="SHIFT-JIS") for i  in df_name]
df_lst = [pd.read_csv(f"data3_csv/{i}_new.csv", index_col=0) for i  in df_name]

df_dic = dict(zip(df_name, df_lst)) 
```

```python
a = Analyzer(token_filters=[CompoundNounFilter()])

w_y_lst = []

for y in df_name:
    df_temp = df_dic[y]
    w_y_lst_temp = []
    for idx in range(len(df_temp)):
        txt_temp = df_temp["name"][idx]
        word_lst = []
        tokens = a.analyze(txt_temp)
        for token in tokens:
            word_temp = token.surface
            word_lst.append(word_temp)
        w_y_lst_temp.append(word_lst)
    w_y_lst.append(w_y_lst_temp)
    
word_dic1 = dict(zip(df_name, w_y_lst)) 
```

```python
word_dic1 = dict(zip(df_name, w_y_lst)) 
```

```python
a = []

for year in df_name:
    
    list_temp = word_dic1[year]
    
    join_list = []
    
    for tex in list_temp:
        join_list += tex
        
    a.append(join_list)
    
word_dic2 = dict(zip(df_name, a)) 
```

```python
lst = []

for y in df_name:
    df_temp = pd.DataFrame()
    df_temp["word"] = word_dic2[y]
    df_temp = df_temp[df_temp["word"].str.len() != 1]
    lst_temp = df_temp["word"].tolist()
    
    cc = collections.Counter(lst_temp)
    word, count = zip(*cc.most_common())
    df_temp2 = pd.DataFrame()
    df_temp2["count"] = count
    df_temp2["word"] = word
    
    lst.append(df_temp2)
    
word_dic3 = dict(zip(df_name, lst)) 
```

```python
word_lst = [word_dic3[y]["word"].tolist() for y in df_name]

all_word = []

for lst in word_lst:
    all_word += lst
    
cc = collections.Counter(all_word)
word, count = zip(*cc.most_common())    

word_set = list(set(word))
# word_set = list(set(all_word))

len(word_set)

df_zero = np.zeros((len(word_set),6))

df_count = pd.DataFrame(df_zero)
df_count = df_count.set_axis(word_set,axis="index")
df_count = df_count.set_axis(df_name,axis="columns")
df_count
```

```python
w_n = df_count.index.tolist()

for y in df_name:
    count_temp = word_dic3[y]["count"].tolist()
    word_temp = word_dic3[y]["word"].tolist()
    
    for w in w_n:
        if w in word_temp:
            count_idx = word_temp.index(w)
            df_count.loc[w,y] = count_temp[count_idx]
```
